```python
import pandas as pd
```

```python
arabica = pd.read_csv('data/arabica_data_cleaned.csv')
arabica.columns
```

```python
arabica_features = arabica[['Species','Country.of.Origin','Region','Company','Harvest.Year','Aroma','Cupper.Points','Sweetness','Clean.Cup','Uniformity','Balance','Body','Acidity','Aftertaste','Flavor','Total.Cup.Points']]
arabica_features.head()
```

```python
arabica_features = arabica_features.rename(columns={'Country.of.Origin':'Country','Harvest.Year':'Harvest Year','Cupper.Points':'Cupper Points','Clean.Cup':'Clean Cup','Total.Cup.Points':'Total Cup Points'})
arabica_features.head()
```

```python
robusta = pd.read_csv('data/robusta_data_cleaned.csv')
robusta.columns
```

```python
robusta.columns
```

```python
robusta_features = robusta[['Species','Country.of.Origin','Region','Company','Harvest.Year','Fragrance...Aroma','Cupper.Points','Bitter...Sweet','Clean.Cup','Uniform.Cup','Balance','Mouthfeel','Salt...Acid','Aftertaste','Flavor','Total.Cup.Points']]
robusta_features.head()
```

```python
robusta_features.columns
```

```python
robusta_features = robusta_features.rename(columns={'Country.of.Origin':'Country','Harvest.Year':'Harvest Year','Fragrance...Aroma':'Aroma','Cupper.Points':'Cupper Points','Bitter...Sweet':'Sweetness','Clean.Cup':'Clean Cup',
                                                   'Uniform.Cup':'Uniformity','Mouthfeel':'Body','Salt...Acid':'Acidity','Total.Cup.Points':'Total Cup Points'})
robusta_features.head()
```

```python
arabica_features.head()
```

```python
combined_df = pd.concat([arabica_features,robusta_features])
```

```python
combined_df.columns = map(str.lower, combined_df.columns)
combined_df.columns = combined_df.columns.str.replace(' ', '_')
```

```python
combined_df.columns
```

```python
combined_df.to_csv('coffeeclean.csv', index=False,encoding='utf-8')
```

```python
coffee_df = pd.read_csv('coffeeclean.csv')
coffee_df.head()
```

```python
coffeequality = coffee_df.drop(['region','company'], axis=1)
coffeequality.dtypes
```

```python
coffeequality['country']=coffeequality['country'].replace(r"\(.*\)","")
coffeequality['country']=coffeequality['country'].apply(str)
```

```python
# function to remove non-ASCII 
def remove_non_ascii(country): 
    return ''.join(i for i in country if ord(i)<128) 
 
coffeequality['country'] = coffeequality['country'].apply(remove_non_ascii)
```

```python
coffeequality.count()
```

```python
coffeequality.to_csv('coffeequality.csv', index=False,encoding='utf-8')
```
